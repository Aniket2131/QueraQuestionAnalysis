```python
import re
```

```python
def textPreProcess(que):
    
    que = str(que).lower().strip()
    
    # replace special character with string equivalent
    que = que.replace('%', 'percent')
    que = que.replace('$', 'dollar')
    que = que.replace('₹', 'rupee')
    que = que.replace('€', 'euro')
    que = que.replace('@', 'at')
    
    # pattern '[Math]' appears 900 times is the dataset
    que = que.replace('[Math]', '')
    
    # replacing some number with string
    que = que.replace(',000,000,000', 'b')
    que = que.replace(',000,000', 'm')
    que = que.replace(',000', 'k')
    que = re.sub(r'([0-9]+)000000000', r'\1b', que)
    que = re.sub(r'([0-9]+)000000', r'\1m', que)
    que = re.sub(r'([0-9]+)000', r'\1k', que)
    
    # Contraction english words
    contraction = {
        "aren't": "are not",
        "there's": "there is",
        "can't": "can not",
        "they'd": "they had",
        "couldn't": "could not",
        "they'll": "they will",
        "didn't": "did not",
        "they're": "they are",
        "doesn't": "does not",
        "they've": "they have",
        "don't": "do not",
        "we'd": "we had",
        "hadn't": "had not",
        "we're": "we are",
        "hasn't": "has not",
        "we've": "we have",
        "haven't": "have not",
        "weren't": "were not",
        "he'd": "he had",
        "what'll": "what will",
        "he'll": "he will",
        "what're": "what are",
        "he's": "he is",
        "what's": "what is",
        "i'd": "i had",
        "what've": "what have",
        "i'll": "i will",
        "where's": "where is",
        "i'm": "i am",
        "who'd": "who had",
        "i've": "i have",
        "who'll": "who will",
        "isn't": "is not",
        "who'e": "who are",
        "let's": "let us",
        "who's": "who is",
        "mightn't": "might not",
        "who've": "who have",
        "mustn't": "must not",
        "won't": "will not",
        "shan't": "shall not",
        "wouldn't": "would not",
        "she'd": "she had",
        "you'd": "you had",
        "she'll": "she will",
        "you'll": "you will",
        "she's": "she is",
        "you're": "you are",
        "shouldn't": "should not",
        "you've": "you have",
    }
    
    q_decontracted = []
    for word in que.split():
        if word in contraction:
            word = contraction[word]
        q_decontracted.append(word)
    
    que = ' '.join(q_decontracted)
    que = que.replace("'ve", " have")
    que = que.replace("n't", " not")
    que = que.replace("'re'", " are")
    que = que.replace("'ll'", " will")
    
    
    # removing html tags
    que = BeautifulSoup(que).text
    
    
    # removing punctuations
    pat = re.compile('\W')
    que = re.sub(pat, ' ', que).strip()
    
    return que 
```

```python
from nltk.corpus import stopwords
```

```python
def token_features(row):
    
    q1 = row['question1']
    q2 = row['question2']
    
    div = 0.001
    
    stop_words = stopwords.words('english')
    
    all_token_features = [0.0]*8
    
    # question to tokens
    q1_token = q1.split()
    q2_token = q2.split()
    
    if len(q1_token) == 0 or len(q2_token) == 0:
        return all_token_features
    
    # question to non_stop_words/words
    q1_words = set([word for word in q1_token if word not in stop_words])
    q2_words = set([word for word in q2_token if word not in stop_words])
    
    # question to stop_words
    q1_stopword = set([word for word in q1_token if word in stop_words])
    q2_stopword = set([word for word in q2_token if word in stop_words])
    
    # common token from token set
    common_token = len(set(q1_token).intersection(set(q2_token)))
    
    # common words/non-stopwords
    common_word = len(q1_words.intersection(q2_words))
    
    # common stopwords
    common_stopwords = len(q1_stopword.intersection(q2_stopword))
    
    # adding all the token features
    
    all_token_features[0] = common_word/(min(len(q1_words), len(q2_words)) + div)
    all_token_features[1] = common_word/(max(len(q1_words), len(q2_words)) + div)
    all_token_features[2] = common_stopwords/(min(len(q1_stopword), len(q2_stopword)) + div)
    all_token_features[3] = common_stopwords/(max(len(q1_stopword), len(q2_stopword)) + div)
    all_token_features[4] = common_token/(min(len(q1_token), len(q2_token)) + div)
    all_token_features[5] = common_token/(max(len(q1_token), len(q2_token)) + div)
    
    # last word equal or not
    all_token_features[6] = int(q1_token[-1] == q2_token[-1])
    
    # fist word equal or not 
    all_token_features[7] = int(q1_token[0] == q2_token[0])
    
    return all_token_features
```

```python
token_features = new_df9.apply(token_features, axis=1)
```

```python
new_df9['cwc_min'] = list(map(lambda x: x[0], token_features))
new_df9['cwc_max'] = list(map(lambda x: x[1], token_features))
new_df9['csc_min'] = list(map(lambda x: x[2], token_features))
new_df9['csc_max'] = list(map(lambda x: x[3], token_features))
new_df9['ctc_min'] = list(map(lambda x: x[4], token_features))
new_df9['ctc_max'] = list(map(lambda x: x[5], token_features))
new_df9['last_word_equal'] = list(map(lambda x: x[6], token_features))
new_df9['first_word_equal'] = list(map(lambda x: x[7], token_features))
```

```python
def length_features(row):
    
    q1 = row['question1']
    q2 = row['question2']
    
    all_length_features = [0.0]*3
    
    # question to tokens
    q1_token = q1.split()
    q2_token = q2.split()
    
    if len(q1_token) == 0 or len(q2_token) == 0:
        return all_length_features
    
    # mean length
    all_length_features[0] = (len(q1_token)+len(q2_token))/2
    
    # abs length 
    all_length_features[1] = abs(len(q1_token)-len(q2_token))
    
    # longest substring ratio
    strg = list(distance.lcsubstrings(q1, q2)) 
    all_length_features[2] = len(strg[0])/(min(len(q1), len(q2)) + 1)
    
    return all_length_features
```

```python
length_features = new_df9.apply(length_features, axis=1)
```

```python
new_df9['mean_len'] = list(map(lambda x: x[0], length_features))
new_df9['abs_len'] = list(map(lambda x: x[1], length_features))
new_df9['lcs_ratio'] = list(map(lambda x: x[2], length_features))
new_df9.sample()
```

```python
def fuzzy_feature(row):
    
    q1 = row['question1']
    q2 = row['question2']
    
    fuzzy = [0.0]*4
    
    fuzzy[0] = fuzz.ratio(q1, q2)
    fuzzy[1] = fuzz.partial_ratio(q1, q2)
    fuzzy[2] = fuzz.token_sort_ratio(q1, q2)
    fuzzy[3] = fuzz.token_set_ratio(q1, q2)
    
    return fuzzy 
```

```python
fuzzy = new_df9.apply(fuzzy_feature, axis=1)
```

```python
new_df9['fuzzy_ratio'] = list(map(lambda x: x[0], fuzzy))
new_df9['fuzzy_partial_ratio'] = list(map(lambda x: x[1], fuzzy))
new_df9['fuzzy_sort_ratio'] = list(map(lambda x: x[2], fuzzy))
new_df9['fuzzy_set_ratio'] = list(map(lambda x: x[3], fuzzy))
new_df9.sample()
```
